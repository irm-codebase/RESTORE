```python
import pandas as pd
import numpy as np
```

```python
path = "data/sources/CHE/Transport/passenger/road_newstock_90_21.csv"
new = pd.read_csv(path, index_col=[0,1,2,3], header=0)
new = new.loc["Switzerland"]
path = "data/sources/CHE/Transport/passenger/road_totalstock_90_21.csv"
total = pd.read_csv(path, index_col=[0,1,2,3], header=0)
total = total.loc["Switzerland"]
```

```python
group = "> Motorcycles"

lmv = new.xs(group)
fuel_types = lmv.index.unique(level="Fuel")
new_by_t_y = pd.DataFrame()
for i in fuel_types:
    fuel = lmv.xs(i, level="Fuel")
    new_by_t_y[i] = fuel.sum()

lmv = total.xs(group)
fuel_types = lmv.index.unique(level="Fuel")
tot_by_t_y = pd.DataFrame()
for i in fuel_types:
    fuel = lmv.xs(i, level="Fuel")
    tot_by_t_y[i] = fuel.sum()
```
